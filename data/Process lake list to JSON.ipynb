```python
import pandas as pd
import json
```

```python
df = pd.read_csv("lakes.csv")
```

```python
df
```

```python
out = []
for index, row in df.iterrows():
    out.append({"id": row["id_int"], "name": row["Name"], "coords": [round(float(row["repr_point_lat"]),3), round(float(row["repr_point_lon"]),3)]})
```

```python
out
```

```python
with open('lakes.json', 'w') as fp:
    json.dump(out, fp)
```
