```python
import pandas as pd
import json
```

```python
df = pd.read_csv("lakes.csv")
df[['letter', 'number', 'thirdcol']] = df['id_str'].str.extract('([A-Za-z]+)(\d+\.?\d*)([A-Za-z]*)', expand = True)
df["number"] = df["number"].str.lstrip("0")
df["id_str2"] = df["letter"]+df["number"]
```

```python
df
```

```python
out = []
for index, row in df.iterrows():
    out.append({"id": row["id_int"], "id_str":row["id_str2"], "name": row["Name"], "coords": [round(float(row["repr_point_lat"]),3), round(float(row["repr_point_lon"]),3)]})
```

```python
out
```

```python
with open('lakes.json', 'w') as fp:
    json.dump(out, fp)
```
